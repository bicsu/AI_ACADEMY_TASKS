```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules
from mlxtend.frequent_patterns import apriori
```

```python
ds_asso = [["소주",'콜라', '맥주'],
          ['소주','콜라','와인'],
          ['소주', '주스',],
          ['콜라','맥주',],
           ['소주','콜라','맥주','와인',],
           ['주스']]
```

```python
enc = TransactionEncoder()
arr = enc.fit_transform(X=ds_asso)
ds_pre = pd.DataFrame(arr, columns= enc.columns_)
ds_pre
```

```python
min_support = 0.5
ds_freq = apriori(ds_pre, min_support = 0.5, use_colnames = True)

ds_freq.round(3)
```

```python
# 기준 = 신뢰도, 최소 신뢰도 = 0.5
ds_asso_rule = association_rules(ds_freq, metric = "confidence", min_threshold=0.5)
ds_asso_rule.round(3)
```

* [(맥주), (콜라)] Confidence = 1.0 : 맥주를 구매한 고객은 콜라도 반드시 구매
* [(콜라), (맥주)] Confidence = 0.75 : 콜라를 구매한 고객 중 맥주를 고른 고객은 0.75
* [(콜라), (소주)] Lift = 1.125 [(콜라), (맥주)] Lift = 1.5 : 콜라를 구매한 고객들에게는 소주보다 맥주를 추천

```python
# 기준 = 신뢰도, 최소 신뢰도 = 0.5
ds_asso_rule = association_rules(ds_freq, metric='confidence', min_threshold=0.5)
ds_asso_rule.round(3)
```
