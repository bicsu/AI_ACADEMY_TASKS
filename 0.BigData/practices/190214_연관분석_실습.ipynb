```python
import pandas as pd
data = pd.read_csv("./csvfiles/상품구매.csv")
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules
from mlxtend.frequent_patterns import apriori
```

```python
ID = list(set(data["ID"]))
ID.sort()

list_association = []
for i in ID:
    temp_list = list(data[data["ID"]==i]["PRODUCT"])
    temp_list.sort()
    list_association.append(temp_list)
    
for row in list_association :
    print(row)
```

```python
enc = TransactionEncoder()
arr = enc.fit_transform(X=list_association)
ds_pre = pd.DataFrame(arr, columns= enc.columns_)
ds_pre.head()
```

```python
min_support = 0.5
ds_freq = apriori(ds_pre, min_support = 0.5, use_colnames = True)

ds_freq
```

```python
# 기준 = 신뢰도, 최소 신뢰도 = 0.5
ds_asso_rule = association_rules(ds_freq, metric = "confidence", min_threshold=0.5)
ds_asso_rule.round(3)
```
